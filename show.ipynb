# Fig.4

```python
# The result of N=9 could be 8 or 9 because the partition with N is not unique and our realization has random results.
g,_ = graph('dolphins')
array = [0,3,4,5,6,7,8,9]
regions_dict_l = get_partition(g, array, [0])
regions_dict_N = get_partition(g, [0], array)

max_size_l = [max([len(region) for region in regions_dict_l[(l,0)]]) for l in array]
max_size_n = [max([len(region) for region in regions_dict_N[(0,n)]]) for n in array]
array[0] = 2
# print(max_size_l[1:], max_size_n[1:], max_size_l[0], sep='\n')
print('max_size_exact:', max_size_l[0])
print('max_size_to_l:', max_size_l[1:])
print('max_size_to_n:', max_size_n[1:])

plt.figure(1,figsize = (3.375,2.5))
plt.plot(array, [2] + max_size_l[1:], label='$L$', marker='o')
plt.plot(array, [2] + max_size_n[1:], label='$N$', marker='x')
plt.xlim(2,9)
plt.ylim(0,60)
plt.ylabel('Size of largest region')
plt.xlabel('Value of $L/N$')
plt.legend()
# plt.savefig('region_size.pdf',bbox_inches='tight')
```
